```python
df_office_episodes = episode[episode['parentTconst'] == 'tt0386676']
```

```python
df_office_episodes
```

```python
ratings[ratings['tconst'] == 'tt0664510']
```

```python
df_office_ratings = df_office_episodes.merge(ratings, how='left', on='tconst')
```

```python
df_office_ratings['seasonNumber'] = df_office_ratings['seasonNumber'].astype(int)
df_office_ratings['episodeNumber'] = df_office_ratings['episodeNumber'].astype(int)
```

```python
df_office_ratings.info()
```

```python
df_final = df.merge(df_office_ratings, how='inner', left_on=['season', 'episode'], right_on = ['seasonNumber','episodeNumber'])
```

```python
df_final = df_final[['index', 'season','episode','episode_name','director','writer','character','text','averageRating','numVotes']]
```

```python
df_final.to_csv('data/final_office.csv')
```
